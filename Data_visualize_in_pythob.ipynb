```
import numpy as np
import matplotlib.pyplot as plt
import statistics 

Roll=[x for x in range(1,41)]
marks=np.random.randint(0,30,40)
plt.plot((0,40),(np.mean(marks),np.mean(marks)),linestyle='dotted',label="Mean")
plt.plot((0,40),(np.median(marks),np.median(marks)),linestyle='dotted',label="Median")
#plt.plot((0,40),(statistics.stdev(marks),statistics.stdev(marks)),linestyle='--',label="Standard deviation")
plt.plot(Roll,marks,marker='o',label="Data")
plt.title("Term test result")
plt.xlabel("Roll")
plt.ylabel("Marks")
plt.legend()
plt.show()

 
```
